# Q2. Which sport is having most number of Gold Medals so far? (Top5)

```python
medals = pd.get_dummies(df["Medal"])  
data = pd.concat([df , medals], axis = 1 )
```

```python
df.head()
```

```python
df = pd.DataFrame({"Sport":data["Sport"],"Gold":data["Gold"]})
df.groupby("Sport").sum()["Gold"].sort_values(ascending = False).head()
```

```python
v1 = df.groupby("Sport").sum()["Gold"].sort_values(ascending = False).head()
v1.plot(kind = "bar" , figsize = (10,10), ylabel = "Gold Medals" , title = "Top 5 Sports with most no. of Gold Medals")
```
